```python
from bs4 import BeautifulSoup
import urllib.request

# https://www.glamuse.com/pyjama.html
brand = []
product = []
original_price = []
price = []
product_link = []

for i in range(1,4):
    
    url = 'https://www.glamuse.com/pyjama.html?page=' + str(i)

# Use urllib2 module to open the url 
    ourUrl = urllib.request.urlopen(url)

# create a BeautifulSoup object which represents the document as a nested data structure
    soup = BeautifulSoup(ourUrl,'html.parser')
    for i in soup.find_all('div',{'class':'div_listings'}):
        if i.find('a',{'class':'nom_produit'}) != None:
            brand.append(i.find('a',{'class':'nom_produit'}).span)
        if i.find('div',{'class':'listings_nom_produit_sub'}) != None:
            product.append(i.find('div',{'class':'listings_nom_produit_sub'}).text)
        
        
        if str(i.find('div',{'class':'listings_prix_produit_ht'})).find('itemprop="url"')>0:
            if i.find('div',{'itemprop':'offers'}).find('div',{'class':'prix_barre_listing_rose'}) != None:
                original_price.append(i.find('div',{'itemprop':'offers'}).find('div',{'class':'prix_barre_listing_rose'}).text)
            else: 
                original_price.append(None)
            
        if i.find('a',{'itemprop':'url'}) != None:
            price.append(i.find('a',{'itemprop':'url'}).text)
            product_link.append(i.find('a',{'itemprop':'url'}).get('href'))
```

```python
pic = []
for i in product_link:
    url = i

    # Use urllib2 module to open the url 
    ourUrl = urllib.request.urlopen(url)

    # create a BeautifulSoup object which represents the document as a nested data structure
    soup = BeautifulSoup(ourUrl,'html.parser')

    for i in soup.find_all('div',{'class':'panel carousel-cell img_principale'}):
        index = str(i.img.get('data-cloudzoom')).find('zoomPosition')
        pic.append(str(i.img.get('data-cloudzoom'))[11:index-3])
```

```python
import os
import requests as req

os.mkdir('photos')
i = 1
```

```python
for index, link in enumerate(pic):
    img_data = req.get(link).content
    with open('photos/'+str(index+1)+'.png','wb+') as f:
        f.write(img_data)
    i+=1
```

```python
import pandas as pd

column_names = ["brand", "product", "original price", 'price', 'product link', 'picture link']
df = pd.DataFrame(columns = column_names)

df['brand'] = brand
df['product'] = product
df['original price'] = original_price
df['price'] = price
df['product link'] = product_link
df['picture link'] = pic
df.to_csv('glamuse.csv')
```
